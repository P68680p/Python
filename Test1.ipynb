```
# 5.Циклы. Задание 3
#По данному натуральному n ≤ 9 выведите лесенку из n ступенек, i-я ступенька состоит из чисел от 1 до i без пробелов.
#Разрешается использовать функцию str(число). Данная функция переводит числовые данные в строковые.

v= int(input('Введите число ступеней: \n'))
s=""
for i in range(1,10):
  if i==v+1:
    break
  else:
    z=str(i)
    s+=z
    print(s)
```

```
# 5.Циклы. Задание 4
#По данному натуральному n ≤ 9 выведите лесенку из n ступенек, i-я ступенька состоит из чисел от 1 до i без пробелов.
#Разрешается использовать функцию str(число). Данная функция переводит числовые данные в строковые.
#Дополните код из предыдущего задания, чтобы теперь получалась пирамида. То есть каждая ступень состоит из чисел от 1 до i и обратно.
v= int(input('Введите число ступеней: \n'))
s=""
s1=""
for i in range(1,10):
  if i==v+1:
    break
  else:
    z=str(i)
    s1=s
    s+=z
    print((v-i)*" "+s+s1[::-1]+(v-i)*" ")
```
